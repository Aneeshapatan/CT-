```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Ensure column is properly formatted
df['nic_name'] = df['nic_name'].astype(str).fillna("")

def clean_and_tokenize(text):
    text = text.lower()  # Convert to lowercase
    text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation
    tokens = word_tokenize(text)  # Tokenization
    stop_words = set(stopwords.words('english'))  # Load stopwords
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    return tokens

# Apply function to the column
df['nic_name'] = df['nic_name'].apply(clean_and_tokenize)
```

```python
from gensim.models import Word2Vec

# Assuming 'cleaned_tokens' is the column containing tokenized text
sentences = df['nic_name'].tolist()

# Train Word2Vec model
model = Word2Vec(df['nic_name'], vector_size=100, window=5, min_count=1, workers=4)

# Save the Word2Vec model
model.save("word2vec_model")

# You can use the trained Word2Vec model to get word vectors
word_to_check = 'uranium'
word_vector = model.wv[word_to_check]
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

# Load your dataset
df = pd.read_csv('final_HR.csv')  # Update the filename

# Initialize geolocator
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to get latitude and longitude
def get_lat_lon(state, district):
    try:
        location = geolocator.geocode(f"{district}, {state}, India")
        if location:
            return location.latitude, location.longitude
        else:
            print(f"Location not found for {district}, {state}")
            return None, None
    except Exception as e:
        print(f"Error for {district}, {state}: {e}")
        return None, None

# Add latitude and longitude columns
df['latitude'] = None
df['longitude'] = None

# Loop through each district and get coordinates
for index, row in df.iterrows():
    state = row['State']
    district = row['District']
    lat, lon = get_lat_lon(state, district)
    df.at[index, 'latitude'] = lat
    df.at[index, 'longitude'] = lon
    print(f"Processed {district}, {state} -> Lat: {lat}, Lon: {lon}")
    time.sleep(1)  # To respect API rate limits

# Save the updated CSV
df.to_csv('final_HR_with_latlon.csv', index=False)
print("Latitude and Longitude added successfully!")
```
